**Select only origin, dest and tailnum from flights dataset. Remove flights without a tailnum**

```r
flights_short <- flights %>% select(origin, dest,tailnum) %>% filter(!is.na(tailnum))
```

```r
flights_short
```

**Create a short version of planes dataset. Remove records with NA seats**

```r
planes_short <- planes %>% select(tailnum,seats) %>% filter(!is.na(seats))
```

```r
planes_short
```

**Now we will create a joined table that shows seat information for the flights**

```r
flights_with_seats <- flights_short %>% left_join(planes_short,by=c("tailnum"="tailnum"))
```

```r
flights_with_seats
```

**Create a summary of flight seats where data is grouped by origin and destination. The summary column is average seats**

```r
flights_with_seats_long <- flights_with_seats %>% group_by(origin,dest) %>% summarize(avg_seats = mean(seats,na.rm=T))
```

```r
flights_with_seats_long
```

**Now make the long average seats table wide using pivot_wider**

```r
flights_with_seats_wide <- flights_with_seats_long %>% pivot_wider(id_cols=dest, names_from=origin, values_from=avg_seats)
```

```r
flights_with_seats_wide
```

**Now let's go back to long form**

```r
flights_with_seats_long_again <- flights_with_seats_wide %>% pivot_longer(cols=-"dest",names_to="origin", values_to="avg_seats")
```

```r
flights_with_seats_long_again
```

**I would like to display avg seats per origin and destination sorter descending avg_seats**

```r
flights_with_seats_long_again %>% arrange(-avg_seats)
```

```r
weather_with_org_yr_mt_dy <- weather %>% unite(col="weather_key",c("origin","year","month","day"))
```
